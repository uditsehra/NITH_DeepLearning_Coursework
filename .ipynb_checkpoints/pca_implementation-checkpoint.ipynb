# Principal Component Analysis Implementation on meaningful data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
data.keys()

# data['target_names']
```

```python
data['feature_names']
```

#### As we can see there are many features to consider, so we can apply PCA to reduce the dimensionality and retain the same ammount of information

```python
dataframe = pd.DataFrame(data['data'], columns=data['feature_names'])

scaler = StandardScaler()
scaler.fit(dataframe)
```

```python
dataframe
```

```python
scaled_data = scaler.transform(dataframe)
```

```python
pca = PCA(n_components=5)
pca.fit(scaled_data)
x = pca.transform(scaled_data)
```

```python
x.shape
```

```python
pca.components_
```

#### Visualise these components

```python
data['target']
```

```python
plt.figure(figsize=(10,10))
pc1 = x[:,0]
pc2 = x[:,1]
# plt.scatter(pc1, pc2, c=data['target_names'], cmap='plasma')
plt.scatter(x[:, 0],x[:, 1],c=data['target'],cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
```

#### For three principal components we need to visualise this using a 3D graph more than that won't be feasible

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))

axis = fig.add_subplot(111, projection='3d')

axis.scatter(x[:,0],x[:,1],x[:,2], c=data['target'], cmap='plasma')
axis.set_xlabel("PC1", fontsize=10)
axis.set_ylabel("PC2", fontsize=10)
axis.set_zlabel("PC3", fontsize=10)
```
